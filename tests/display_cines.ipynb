```python
import cv2
import IPython
import os

from pycine.color import color_pipeline, resize
from pycine.raw import read_frames

def imshow_save(img):
    _,ret = cv2.imencode('.jpg', img) 
    i = IPython.display.Image(data=ret)
    IPython.display.display(i)
    with open('test.jpg', 'wb') as f:
        f.write(ret)

def display_frames(cine_file, start_frame=0, count=1):
    raw_images, setup, bpp = read_frames(cine_file, start_frame=start_frame, count=count)

    for i, img in enumerate(raw_images):
        frame = start_frame + i


        imshow_save(img)
    
    print(frame)
display_frames("../data/raw_cine/Ar-21-5Mpa-105Mpa-9ms-2.cine", count=1)
```
